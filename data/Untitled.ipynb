```python
plt.scatter(listings['price'], listings['longitude'])
plt.ylabel('price')
plt.xlabel('longitude')
```

```python
Y = listings['price']
X = listings[['longitude']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.4, shuffle=True)
```

```python
sns.scatterplot(x=X_train['longitude'], y=y_train, label='train')
sns.scatterplot(x=X_test['longitude'], y=y_test, label='test')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
print('alpha is',lr.intercept_)
print('beta is', lr.coef_)
```

```python
lr.predict(X_test)
```

```python
sns.scatterplot(x=X_train['longitude'], y=y_train, label='train')
sns.scatterplot(x=X_test['longitude'], y=y_test, label='test')

x_range = np.arange(start=listings['longitude'].min(),stop=listings['longitude'].max())
y_hat = lr.intercept_ + lr.coef_[0]*x_range
sns.lineplot(x=x_range,y=y_hat,color="red")
```
